```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
readData = pd.read_csv("amazon.csv")
```

```python
# Displaying the sample Data
display(readData.head())
```

```python
readData.dtypes
```

```python
date = readData["Date"]
openingValue = readData["Open"]
closingValue = readData["Close"]
```

```python
#Line Plot
plt.figure(figsize=(15, 10))
plt.plot(date,openingValue,marker=".",mec="r",mfc="r",label="Opening Value")
plt.plot(date,closingValue,c="r",marker=".",mec="b",mfc="b",label="Closing Value")
plt.legend()
font = {'color':'blue','size':'20'}
plt.title("Opening and Closing stock value of Amazon in August Month 2022",fontdict=font)
plt.xlabel("Date",fontdict=font)
plt.ylabel("Opening and Closing price in ($)",fontdict=font)
plt.show()
plt.close()
```
